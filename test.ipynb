```python
import torch

t = torch.rand(1, 1, 5, 5) * 10
s = torch.rand(1, 1, 5, 5) * 5

v = s * t
print("v:", v > 1)
```

```python
t = torch.nn.functional.softmax(t, dim=2)
s = torch.nn.functional.softmax(s, dim=2)

v = s * t
print("v after softmax:", v > 1)
print("v after softmax:", torch.sum(v, dim=3))

```
